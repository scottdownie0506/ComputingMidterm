```python
# question 1

import matplotlib.pyplot as plt
import numpy as np
import json

A = [2,4,6,8,4,5,2,1,9,0,4,6,7,4,3,2,1,9,10,3,7,9,6,0,1,3,5,6,7,8,9,10,2,3,6,8,9,10,6,7,4,3]
B = sorted(A)

A_count = {}

for number in B:
    if number in A_count:
        A_count[number] += 1
    else:
        A_count[number] = 1
        
for k, v in A_count.items():
   print(f"{k} : {v}")

print()

for k, v in A_count.items():
    print(k, 'occurs', v, 'times in list A.')

print()

A_count_keys = list(A_count.keys())
A_count_values = list(A_count.values())
x = np.array(A_count_keys)
y = np.array(A_count_values)

plt.bar(x, y, color='r')
plt.title('Frequencies of Numbers in a List')
plt.xlabel('Numbers')
plt.ylabel('Frequency')

plt.show()

with open("question1.json", "w") as outfile:
    json.dump(A_count, outfile)
```

```python
# loading data frame and changing data to fit my information 
import pandas as pd

df = pd.read_csv('amazon-dataset.csv')

df['Ordering Customer Email'] = 'scottdownie0506@gwu.edu'
df['Shipping Address Name'] = 'Scott Downie'
df['Shipping Address Street 1'] = '2121 H St NW'
df['Buyer Name'] = 'Scott Downie'

df
```

```python
df.loc[:, ['Total Charged']] = df['Total Charged'].str.replace("$", '').astype('float')
```

```python
df.loc[:, ['Shipping Charge']] = df['Shipping Charge'].str.replace("$", '').astype('float')
```

```python
df.loc[:, ['Tax Charged']] = df['Tax Charged'].str.replace("$", '').astype('float')
```

```python
df.loc[:, ['Total Promotions']] = df['Total Promotions'].str.replace("$", '').astype('float')
```

```python
total_avg = df['Total Charged'].mean()
total_max = df['Total Charged'].max()
total_min = df['Total Charged'].min()

ship_avg = df['Shipping Charge'].mean()
ship_mode = float(df['Shipping Charge'].mode())
ship_max = df['Shipping Charge'].max()

promo_max = df['Total Promotions'].max()

tax_avg = df['Tax Charged'].mean()
tax_max = df['Tax Charged'].max()
```

```python
print('On average, I spent', total_avg, 'on each order.')
print('The most I spent on a single order was', total_max)
print('The least I spent on an order was', total_min)
print()
print(ship_mode, 'was the usual cost for shipping.')
print('The most I paid for shipping was', ship_max)
print('But on average, I paid', ship_avg, 'for shipping.')
print()
print('The biggest discount I received on a single order was', promo_max)
```
